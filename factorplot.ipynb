```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
```

```python
g = sns.factorplot(x="day", y="total_bill", hue="sex", data=tips, dodge=True)
```

```python
g = sns.factorplot(x="day", y="total_bill", hue="sex", 
                   data=tips, kind="violin", scale="width")
```

```python
g = sns.factorplot(x="day", y="total_bill", hue="sex", 
    data=tips, kind="violin", scale="count",
    col="time")
```

```python
titanic = sns.load_dataset("titanic")

g = sns.factorplot(x="age", y="embark_town",
                    hue="sex", row="class",
                    data=titanic[titanic.embark_town.notnull()],
                    orient="h", size=2, aspect=3.5, palette="Set3",
                    kind="violin", split=True, cut=0, bw=.2)
```

```python
g = sns.factorplot(x="age", y="embark_town",
                    hue="who", row="class", col="sex",
                    data=titanic[titanic.embark_town.notnull()],
                    orient="h", size=2, aspect=3.5, palette="Set3",
                    kind="box")
```
