```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
```

```python
ax = sns.violinplot(x=tips["total_bill"], inner="box")
```

```python
ax = sns.violinplot(x=tips["total_bill"], inner="quartile")
```

```python
ax = sns.violinplot(x=tips["total_bill"], inner="stick")
```

```python
ax = sns.violinplot(x=tips["total_bill"], inner="point")
```

```python
ax = sns.violinplot(x="day", y="total_bill", data=tips, scale="area")
```

```python
ax = sns.violinplot(x="day", y="total_bill", data=tips, scale="width")
```

```python
ax = sns.violinplot(x="day", y="total_bill", data=tips, scale="count")
```

```python
ax = sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips, palette="muted", scale="count")
```

```python
ax = sns.violinplot(x="day", y="total_bill", hue="smoker",
                    data=tips, palette="muted", split=True)
```
